```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.linear_model import LinearRegression  # Example model, replace as needed

# Step 5: Define the multi-output model
model = MultiOutputRegressor(LinearRegression())  # Replace with your desired model
```

```python
if len(X_train.shape) == 1:
    X_train_reshaped = X_train.reshape(-1, 1)
```

```python
if len(X_test.shape) == 1:
    X_test_reshaped = X_test.reshape(-1, 1)
```

```python
model.fit(X_train_reshaped,y_train)
```

```python
y_Predict = model.predict(X_test_reshaped)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
from math import sqrt
```

```python
print("R2 score  :",r2_score(y_test,y_Predict))
print("MSE score  :",mean_squared_error(y_test,y_Predict))
print("RMSE: ",sqrt(mean_squared_error(y_test,y_Predict)))
```
